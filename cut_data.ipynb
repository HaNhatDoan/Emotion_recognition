```python
import pandas as pd

# Load CSV không header
df = pd.read_csv("fer2013_mp_landmarks.csv", header=None)

# index cột: 0=img_path, 1=subset, 2=emotion_id, 3=emotion_name
subset_col = 1

# Tách train
df_train = df[df[subset_col] == 'train']
df_train.to_csv("fer2013_train.csv", index=False, header=False)
print("Saved train CSV:", df_train.shape)

# Tách test
df_test = df[df[subset_col] == 'test']
df_test.to_csv("fer2013_test.csv", index=False, header=False)
print("Saved test CSV:", df_test.shape)
```
