```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dataset_main/fer2013_mp_landmarks.csv')
```

```python
df = df.rename(columns={'3': 'emotion', '1': 'usage'})

# Kiểm tra lại tên cột
print(df.columns)
```

```python
# Thiết lập phong cách biểu đồ
sns.set_style("whitegrid")

# Tạo khung vẽ
plt.figure(figsize=(10, 6))

# Vẽ biểu đồ cột (Bar chart) đếm số lượng
# order=df['emotion'].value_counts().index giúp sắp xếp cột từ cao đến thấp
ax = sns.countplot(data=df, x='emotion', order=df['emotion'].value_counts().index, palette='viridis')

# Thêm tiêu đề và nhãn
plt.title('Thống kê số lượng mẫu theo từng cảm xúc (Labels)', fontsize=15)
plt.xlabel('Cảm xúc', fontsize=12)
plt.ylabel('Số lượng mẫu', fontsize=12)

# Hiển thị số cụ thể trên đầu mỗi cột
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=11, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.show()

# In ra số liệu cụ thể
print("Số lượng mẫu chi tiết:")
print(df['emotion'].value_counts())
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='usage', palette='pastel')
plt.title('Phân bố dữ liệu theo tập (Train/Test/Val)')
plt.xlabel('Loại tập dữ liệu')
plt.ylabel('Số lượng')
plt.show()
```
